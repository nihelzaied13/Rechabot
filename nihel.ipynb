```python
def get_gpt3_response(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  
        messages=[{"role": "system", "content": "You are a helpful assistant."},
                  {"role": "user", "content": prompt}],
        max_tokens=150,
        temperature=0.7,
    )
    return response.choices[0].message['content'].strip()
```

```python
def recommend_books(category):
    recommendations = books_df[books_df['Category'].str.contains(category, case=False)].head(5)
    if not recommendations.empty:
        return [
            f"Title: {row['Title']}, Author: {row['Authors']}, Publisher: {row['Publisher']}, "
            f"Year: {row['Publish Date (Year)']}, Price: ${row['Price Starting With ($)']}, "
            f"Description: {row['Description']}"
            for _, row in recommendations.iterrows()
        ]
    else:
        return ["No books found in this category."]


def recommend_music(artist_name):
    recommendations = music_df[music_df['Artist Name(s)'].str.contains(artist_name, case=False)].head(5)
    if not recommendations.empty:
        return [
            f"Track: {row['Track Name']}, Artist: {row['Artist Name(s)']}, Album: {row['Album Name']}, "
            f"Release Date: {row['Album Release Date']}, Popularity: {row['Popularity']}"
            for _, row in recommendations.iterrows()
        ]
    else:
        return ["No music found for this artist."]


def recommend_movies(genre):
    recommendations = movies_df[movies_df['listed_in'].str.contains(genre, case=False)].head(5)
    if not recommendations.empty:
        return [
            f"Title: {row['title']}, Director: {row['director']}, Cast: {row['cast']}, "
            f"Country: {row['country']}, Year: {row['release_year']}, "
            f"Rating: {row['rating']}, Description: {row['description']}"
            for _, row in recommendations.iterrows()
        ]
    else:
        return ["No movies found in this genre."]

```

```python
from openai import OpenAI

# Initialize the OpenAI client with your API key
client = OpenAI(
    api_key="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"  # Your actual API key
)

def chat_gpt(prompt):
    # Create a chat completion request
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )
    # Return the response content
    return response.choices[0].message.content.strip()

def get_recommendation(category):
    if category == 'movies':
        prompt = "Do you have a preferred actor, genre, or movie description?"
    elif category == 'books':
        prompt = "Do you have a preferred author, genre, or book description?"
    elif category == 'music':
        prompt = "Do you have a preferred artist or music genre?"
    elif category == 'tv shows':
        prompt = "Do you have a preferred actor, genre, or show description?"
    else:
        prompt = "I'm sorry, I can only provide recommendations for books, music, movies, or TV shows."

    return prompt

def chatbot():
    # Friendly greeting

    greeting_prompt = "Please provide a warm, welcoming greeting for the user, introducing that I can help them with recommendations for books, music, movies, and TV shows."
    greeting = chat_gpt(greeting_prompt)
    print("Assistant:", greeting)

    while True:
        print("What would you like a recommendation for? (Choose from: books, music, movies, tv shows or type 'exit' to quit)")
        category = input("You: ").strip().lower()

        if category in ['exit', 'quit']:
            print("Goodbye! Have a wonderful day!")
            break

        # Ask for further details based on the category
        category_prompt = get_recommendation(category)
        print("Assistant:", category_prompt)

        preference = input("You: ").strip()
        full_prompt = f"Recommend some {category} based on this preference: {preference}"

        # Call the chat_gpt function to get the response
        gpt3_response = chat_gpt(full_prompt)

        # Display the assistant's response
        print("Assistant:", gpt3_response)

if __name__ == "__main__":
    chatbot()
```
